```
# Create a Python dataset (list) that contains the following data:
c1 = ('Zurich',400000,'Limmat')
c2 = ('Vienna',1900000,'Danube')
c3 = ('Paris',2200000,'Seine')
c4 = ('Rome',2900000,'Tiber')
c5 = ('London',8700000,'Thames')

data = [c1,c2,c3,c4,c5]

# Turn the list into a DataFrame
df = sqlContext.createDataFrame(data, ['city','population','river'])

# Show two different ways to select the city on the river Danube 
# DataFrame API
df.filter("river='Danube'").select("city").show()

# SQL
df.registerTempTable("cities")
sqlContext.sql("SELECT city FROM cities WHERE river = 'Danube'").show()
```
